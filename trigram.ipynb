```python
import torch
```

```python
words = open('names.txt', 'r').read().splitlines()
words[:10]
```

```python
chars = sorted(list(set(''.join(words))))
stoi = {s:i+1 for i,s in enumerate(chars)}
stoi['.'] = 0
itos = {i:s for s,i in stoi.items()}
```

```python
N = torch.zeros((28, 28, 28), dtype=torch.int32)
t = {}

for w in words:
    chs = ['.','.'] + list(w) + ['.']
    for ch1, ch2, ch3 in zip(chs, chs[1:], chs[2:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        ix3 = stoi[ch3]
        N[ix1, ix2, ix3] += 1

        trigram = (ch1, ch2, ch3)
        t[trigram] = t.get(trigram, 0) + 1
```

```python
P = (N+0.1).float() # pseudo-count of 1
P /= P.sum((1, 2), keepdim=True)
```

```python
g = torch.Generator().manual_seed(2147483647)

for i in range(15):
    out = []
    ix = 0
    iy = 0
    while True:

        p = P[ix, iy]
        
        ix = iy
        iy = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
        
        if iy == 0:
            break

        out.append(itos[iy])

    print(''.join(out))
```

```python
loglikelihood = 0.0
n = 0

for w in words:
    chs = ['.','.'] + list(w) + ['.']
    for ch1, ch2, ch3 in zip(chs, chs[1:], chs[2:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        ix3 = stoi[ch3]
        prob = P[ix1, ix2, ix3]
        logprob = torch.log(prob)
        n += 1
        loglikelihood += logprob

print(f'{loglikelihood=}')
nll = -loglikelihood
print(f'{nll=}')
print(f'{nll/n=}')
```
